### functions of numpy

### more function
#### like min,max,mean,variance,atandard deviation


```python
a=np.array([[2,1],[3,4]])
print(a)
print(a.min())
print(a.max())
print(np.min(a))
print(np.max(a))
#np.random.seed(a) gives us same value always
print(np.mean(a))

print(np.var(a))
print(np.var(a,axis=0,dtype="int"))
print(np.var(a,axis=1))
print(np.std(a))
print(np.std(a,axis=1))
print(np.std(a,axis=0))
```

```python
#combining two list as one array
a=np.array([1,2,3])
b=np.array([4,5,6])
print(np.stack((a,b)))
print(np.stack((b,a)))
#by giving axis number in stack we can change row as col and col as row
# by default axis is 0
print(np.stack((a,b),axis=1))
```

```python
# dot product matrix multiplication
# cross product vector
n=int(input())
a=[]
for i in range(n):
  temp=input().strip().split()[:n]
  temp=[int(i) for i in temp]
  a.append(temp)
  b=[]
for i in range(n):
  temp=input().strip().split()[:n]
  temp=[int(i) for i in temp]
  b.append(temp)

print(np.dot(a,b))
```

```python
# mean var std
n=[int(i) for i in (input()).strip().split()]
a=[]
for i in range(n[0]):
  temp=input().strip().split()[:n[1]]
  temp=[int(i) for i in temp]
  a.append(temp)
print(np.mean(a,axis=1,dtype="int"))
```

```python
# transpose and flatten matrix
n=[int(i) for i in (input()).strip().split()]
A=[]
for i in range(n[0]):
  temp=input().strip().split()[:n[1]]
  temp=[int(i) for i in temp]
  A.append(temp)
A=np.array(A)
print(np.transpose(A))
print(A.flatten)
```

## pie chart


```python
import numpy as np
import matplotlib.pyplot as plt
labels=["a","b","c","d"]
values=[25,50,35,15]
plt.pie(values,labels=labels)
plt.show()
# explode means gives us separate part brings a out means a explode
plt.pie(values,labels=labels,explode=[0.3,0,0,0])
plt.show()
```
